### What Stat had the highest correlation with chapionship caliber teams?

```python
# create a boxplot for ADJOE: Adjusted Offensive Efficiency 
# (An estimate of the offensive efficiency (points scored per 100 possessions) a team would have against the average Division I defense)
fig = plt.figure(figsize = (10,4))
sns.boxplot(data=Qaulified_For_MM, x='POSTSEASON', y='ADJOE', order=Postseason_labels)
plt.title('Offensive Adjusted Efficiency')
```

```python
#create a boxplot for ADJ_T: Adjusted Tempo (An estimate of the tempo (possessions per 40 minutes) a team would have against the team that wants to play at an average Division I tempo)
fig = plt.figure(figsize = (10,4))
sns.boxplot(data=Qaulified_For_MM, x='POSTSEASON', y='ADJ_T', order=Postseason_labels)
plt.title('Adjusted Tempo (An estimate of the tempo (possessions per 40 minutes)')
```

```python
# create a boxplot for BARTHAG: Power Rating (Chance of beating an average Division I team)
fig = plt.figure(figsize = (10,4))
sns.boxplot(data=Qaulified_For_MM, x='POSTSEASON', y='BARTHAG', order=Postseason_labels)
plt.title('Power Rating (Chance of beating an average Division I team)')
```

### What was the correlation between offensive and defensive stats?

```python
# create 2 heatmaps that show how offensive and defensive stats are correlated if they are
# using seaborne to create these plots and matplot to plot them inline
Offensive_Stats = Qaulified_For_MM[['ADJOE','EFG_O','TOR','ORB','FTR','2P_O','3P_O']]
Defensive_Stats = Qaulified_For_MM[['ADJDE','EFG_D','TORD','DRB', 'FTRD', '2P_D','3P_D']]
Offensive_Corr = Offensive_Stats.corr()
Defensive_Corr = Defensive_Stats.corr()
fig, ax = plt.subplots(1,2, figsize=(20,8))
sns.heatmap(ax=ax[0], data=Offensive_Corr, cmap='magma', annot=True, center=0)
sns.heatmap(ax=ax[1], data=Defensive_Corr, cmap='magma', annot=True, center=0)
ax[0].set_title('Offensive Correlations')
ax[1].set_title('Defensive Correlations')
```

#### Import data and do some basic cleaning

```python
import json
from bs4 import BeautifulSoup
from nltk.tokenize import WordPunctTokenizer
import regex as re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
  
# Opening JSON file
f = open('bball_mm2.txt',)
  
# returns JSON object as
# a dictionary
data = json.load(f)
```

### Tweet Wordcloud

```python
# put the tweets in a list
tweets_list = [x['text'] for x in data]
```

```python
# move the list to a df
df = pd.DataFrame(tweets_list,columns=['Text'])
```

```python
df.head()
```

```python
df.shape
```

```python
#Regex for cleaning
re_list = ['@[A-Za-z0â€“9_]+', '#']
combined_re = re.compile( '|'.join( re_list) )

regex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
```

```python
# Create a function to clean up the tweets and tokenize the words
token = WordPunctTokenizer()
def cleaning_tweets(t):
    del_amp = BeautifulSoup(t, 'lxml')
    del_amp_text = del_amp.get_text()
    del_link_mentions = re.sub(combined_re, '', del_amp_text)
    del_emoticons = re.sub(regex_pattern, '', del_link_mentions)
    lower_case = del_emoticons.lower()
    words = token.tokenize(lower_case)
    result_words = [x for x in words if len(x) > 2]
    return (" ".join(result_words)).strip()
```

```python
# run the function on the tweets to clean it up. 
print("Cleaning the tweets...\n")
cleaned_tweets = []
for i in range(0,2000): #3000 columns in our dataframe
    if( (i+1)%100 == 0 ):
        print("Tweets {} of {} have ben processed".format(i+1,2000))                                                                  
    cleaned_tweets.append(cleaning_tweets((df.Text[i])))
```

```python
# put the data in one string for wordcloud creation
string = pd.Series(cleaned_tweets).str.cat(sep=' ')
```

```python
#get rid of the stopwords to clean up the wordcloud
stopwords = set(STOPWORDS)
stopwords.update(["https", "improveyourlovelifein5words"])
```

```python
# plot the wordcloud
wordcloud = WordCloud(width=1600, stopwords=stopwords,height=800,max_font_size=200,max_words=50,collocations=False, background_color='black').generate(string)
plt.figure(figsize=(40,30))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
